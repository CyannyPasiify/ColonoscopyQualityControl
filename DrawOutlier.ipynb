```python
import json
import copy
```

```python
def keystoint(x):
    try:
        return {int(k):v for k,v in x}
    except ValueError:
        return {k:v for k,v in x}
```

```python
with open('/mnt/data/cwy/ColonoscopyQualityControl/Experiment/R105_detect_fps_vitp14s336c7_400/outliers/outlier_result.json') as json_file:
    data = json.load(json_file,object_pairs_hook=keystoint)
```

```python
data
```

```python
stat = copy.deepcopy(data)
```

```python
max_bin=0
for video in data.keys():
    for label in data[video].keys():
        for situation in data[video][label].keys():
            stat[video][label][situation] = []
            for outlier_item in data[video][label][situation]:
                real_len = outlier_item[1]-outlier_item[0]+1
                if real_len>max_bin:
                    max_bin=real_len
                stat[video][label][situation].append(real_len)
```

```python
stat
```

```python
import seaborn as sns

#形如 stat['NC_KW_inside_00-00-43_00-04-06'][1][0]的数据为该视频的第1个标签的第0种异常情况的异常长度列表
# 现在用要统计每个视频的每个标签的每种异常情况的每个长度的出现次数,并画出柱状图，不能用直方图，因为直方图是连续的，而这里是离散的
lens_stat = dict()
for video in stat.keys():
    for label in stat[video].keys():
        for situation in stat[video][label].keys():
            for lens in stat[video][label][situation]:
                if video not in lens_stat.keys():
                    lens_stat[video] = dict()
                if label not in lens_stat[video].keys():
                    lens_stat[video][label] = dict()
                if situation not in lens_stat[video][label].keys():
                    lens_stat[video][label][situation] = dict()
                if lens not in lens_stat[video][label][situation].keys():
                    lens_stat[video][label][situation][lens] = 0
                lens_stat[video][label][situation][lens] += 1
```

```python
lens_stat
```

```python
label_dict = {0: 'outside',
            1: 'nonsense',
            2: 'ileocecal',
            3: 'bbps0',
            4: 'bbps1',
            5: 'bbps2',
            6: 'bbps3',
              }
```

```python
# 遍历每个视频的每个标签的每种异常情况的每个长度的出现次数，画出对应柱状图，用seaborn
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import os
for video in lens_stat.keys():
    for label in lens_stat[video].keys():
        for situation in lens_stat[video][label].keys():
            x = list(lens_stat[video][label][situation].keys())
            y = list(lens_stat[video][label][situation].values())
            plt.figure(figsize=(8,10))
            plt.bar(x,y)
            # 因为x，y都只有整数，不需要小数位
            # x轴和y轴都只显示整数，比如1，不是1.0
            # 设置其基数为整数1
            
            plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
            plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
            plt.xlabel('Outlier Frame Count')
            plt.ylabel('Outlier Clip Count')
            plt.title(video+'-'+label_dict[label]+'_'+str(situation))
            save_root = '/mnt/data/cwy/ColonoscopyQualityControl/Experiment/R105_detect_fps_vitp14s336c7_400/outliers'
            save_path = os.path.join(save_root,video,video+'-'+label_dict[label]+'_'+str(situation)+'.png')
            if not os.path.exists(os.path.join(save_root,video)):
                os.makedirs(os.path.join(save_root,video))
            # 每个视频一个子文件夹
            plt.savefig(save_path)
            plt.close()
```
